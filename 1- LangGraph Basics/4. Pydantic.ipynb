## Pydantic Data Validation

It is the most widely used data validation library for Python

```python
from langgraph.graph import StateGraph, START, END
from pydantic import BaseModel
```

```python
class State(BaseModel):
    name: str
```

```python
## Node
def example(state: State):
    return {"name" : f"Hello {state.name}"}
```

```python
## State Graph
builder = StateGraph(State)
builder.add_node("example",example)

builder.add_edge(START, "example")
builder.add_edge("example", END)

graph = builder.compile()
```

```python
graph.invoke({"name": "Navneet"})
```
